```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,10)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# airline data
airdf = pd.read_csv('airline_passengers.csv', parse_dates=True, index_col=[0])
```

```python
airdf.dropna(inplace=True)
```

```python
airdf
```

```python
result = seasonal_decompose(airdf['Thousands of Passengers'], model='multiplicative')  # model='mul' also works
result.plot();
```
